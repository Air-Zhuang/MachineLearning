#### We have both countries and clubs to classify players. Let's sum overall score of all the players and find out top countries with highest overall scores.


```python
club = []   #俱乐部名称列表
clubScore = []   #每个俱乐部的Overall数的和
country = []   #国家名称列表
countryScore = []  #每个国家的Overall数的和

for i in df.Nationality.unique():
    country.append(i)
    countryScore.append(df['Overall'][df.Nationality == i].sum())  #筛选当前国家总分之和
    
for i in df.Club.unique():
    club.append(i)
    clubScore.append(df['Overall'][df.Club == i].sum())
    
clubScore = pd.DataFrame({'Club':club,'Score':clubScore}).sort_values('Score',ascending=False)
countryScore = pd.DataFrame({'Country':country,'Score':countryScore}).sort_values('Score',ascending=False)
print(clubScore.head(5))
print(countryScore.head(5))
```

```python
fig,axs = plt.subplots(1,2,figsize=(16,24))
df1 = df.Nationality.value_counts()
sns.set(font_scale=1.3)
sns.barplot(df1[:50].values,df1[:50].index,ax=axs[0],palette='deep').set_title('Top 50 Countries (By Number Of Players)')    #成员最多的50个国家成员数量直方图
sns.barplot(countryScore[:50].Score,countryScore[:50].Country,ax=axs[1],palette='hls').set_title('Top 50 Countries (By Total Score)')  #成员最多的50个国家Overall总和直方图
```

### Here's a simple comparisn between number of players and total of overall scores of countries. Although it is clear that country with more number of players will have more overall score there are countries like japan, china pr , sweden, netherlands, chile and some more who don't follow the trend as shown in the graph. 



#### Let's do same for the clubs and compare them side by side.

```python
fig,axs = plt.subplots(1,2,figsize=(16,24))
df1 = df.Club.value_counts()
sns.set(font_scale=1.3)
sns.barplot(df1[:50].values,df1[:50].index,ax=axs[0],palette='deep').set_title('Top 50 Clubs (By Number Of Players)')   #成员最多的50个俱乐部成员数量直方图
sns.barplot(clubScore[:50].Score,clubScore[:50].Club,ax=axs[1],palette='hls').set_title('Top 50 Clubs (By Total Score)')  #成员最多的50个俱乐部Overall总和直方图
```

### Like countries, there's no such trend as more number of players higher overall score. As seen in first ten Clubs there is no relation in number of players and overall score.
